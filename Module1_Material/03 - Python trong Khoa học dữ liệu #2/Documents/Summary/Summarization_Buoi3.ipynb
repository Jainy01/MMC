# Trực quan trên code

## Khai báo

```python
danh_sach = [5,3,6,3,7,2]
tuplee = (5,3,6,3,7,2)
taphop = {5,3,6,3,7,2}
dictionary = {
    "apple": 5,
    "banana": 2,
    "orange": 3,
    "grape": 1,
    "kiwi": 8,
    "strawberry": 4
}
```

## Thứ tự ?

```python
print(f"Lists: {danh_sach}")
print(f"Tuple: {tuplee}")
print(f"Sets: {taphop} - Có thứ tự")
print(f"Dictionary: {dictionary}")
```

## Chỉ mục?

```python
# Có thể truy xuất bằng chỉ số index
danh_sach[2]
tuplee[2]
```

```python
# Chỉ mục chính là key
dictionary['apple']
```

```python
# Không có chỉ mục
taphop[2]
```

## Có thể sửa đổi?

```python
# List có thể
print(danh_sach)
danh_sach[2] = 100
print(danh_sach)
```

```python
# Tuple không thể, đó là lí do tuple dùng để bảo vệ dữ liệu khi cần
print(tuplee)
tuplee[2] = 100
print(tuplee)
```

```python
# Có thể thay đổi bằng remove, add, nhưng không thể thay đổi tại vị trí cụ thể vì không có chỉ mục
print(taphop)
taphop.remove(3)
taphop.add(8)
print( taphop)
```

```python
print(dictionary)
dictionary["apple"] = 1
print(dictionary)
```

## Có hàm sắp xếp?

```python
print(danh_sach)
danh_sach.sort() # Lưu ý đây là phương thức của list, không phải hàm, nên khi gọi sẽ .sort, khi đó ô nhớ lưu trữ danh sách sẽ sắp xếp lại các giá trị
print(danh_sach)
```

```python
print(tuplee)
tuplee = sorted(tuplee)
print(tuplee)
```

```python
# Thư viện cũng có thể sort theo giá trị - values
footballers_goals = {'Eusebio': 120, 'Cruyff': 104, 'Pele': 150, 'Ronaldo': 132, 'Messi': 125}

sorted_footballers_by_goals = sorted(footballers_goals.items(), key=lambda x:x[1]) # Đây là lamda function, mọi người có thể tìm đọc thêm để hiểu cách sắp xếp này
print(sorted_footballers_by_goals)
```

## Khả năng tìm kiếm

```python
# Cả list và tuple đều có thể dùng hàm index()
ds = [1,2,3,6,2,7]
tup = (5,2,7,2,7,3)
print(ds.index(3))
print(tup.index(5))
```

```python
# Sets không có index nên không thể tìm kiếm bàng index
taphop.index(2)
```

```python
# Dictionary tìm kiếm bằng key
print(dictionary["orange"])
```

## Cách thêm phần tử

```python
my_list = [1, 2, 3]
my_list.append(4)  # Thêm vào cuối danh sách
print("List:", my_list)

my_set = {1, 2, 3}
my_set.add(4)  # Thêm vào danh sách, vị trí tùy vào độ lớn nhỏ của giá trị thêm vào
print("Set:", my_set)

my_tuple = (1, 2, 3)
new_tuple = my_tuple + (4,)  # Bởi vì tuple không thao tác được, muốn thêm chỉ có thể là tạo ra một tuple mới nằm ở ô nhớ khác
print("Tuple:", new_tuple)

my_dict = {"a": 1, "b": 2}
my_dict["c"] = 3  # Thêm bằng key-value hoặc có thể dùng hàm Update()
print("Dictionary:", my_dict)
```
