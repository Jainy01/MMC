# Tổng quan về Markdown

Có lẽ tất cả mọi người đều đã gặp những văn bản tạo bằng markdown nhưng lại không biết đến nó. Ví dụ như những file Readme.md mô tả các repository trên GitHub, hay những phần văn bản mô tả lý thuyết ở trong các notebook code python.   
Markdown chính là một markup language nhẹ phục vụ cho việc soạn thảo văn bản một cách chuyên nghiệp. Cha đẻ của markdown là John Gruder, ông tạo ra ngôn ngữ trên vào năm 2004. Markdown đang là một trong những markup languages được sử dụng phổ biến nhất hiện nay.

## Cách nhanh nhất để chuyển một cell code sang 1 cell markdown:  
* Trên jupyter notebook: Esc - M (Bấm esc để thoát khỏi chế độ gõ, sau đó bấm m để chuyển cell thành markdown cell)
* Trên Google Colab: Ctrl + MM (Nhấn dữ tổ hơn phím theo thứ tự)

## Cách heading, tạo layout

# This is heading 1

## This is heading 2

### This is heading 3

## Các cú pháp thường xuyên sử dụng

* Dùng 2 dấu cách để xuống dòng
* Dùng 2 dấu gạch dưới "_" để set _italic_ style hoặc Hay cặp "**" để set **bold** style
* Dùng cặp "```" để tạo block code trong markdown
  ```
  Print("Hello world")
  ```
* Dùng dấu * để tạo list item
* Hoặc bạn có thể sử dụng các HTML tag như ul, ol để tạo item list:
<ul>
<li>Fish</li>
<li>Eggs</li>
<li>Cheese</li>
</ul>
</ol>
